```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos=np.loadtxt("stellar_radius.dat")
x=[]
y=[]

for i in range(0,datos.shape[0]):
    x.append(datos[i][0])
    y.append(datos[i][1])
    
#Ordenar dos listas


x_list=[]
y_list=[]

for i in range(0,len(x)):
    a=min(x)
    x_list.append(a)
        
    for j in range(0,len(x)):
        if x[j]==a:
            y_list.append(y[j])
            y.remove(y[j])
        
    x.remove(a)
    
#print(x_list)
#print(y_list)


def InterpoladorLineal(x_list,y_list,x):
    for i in range(0,len(x_list)-1):
        if(x>=x_list[i] and x<=x_list[i+1]):
            x_prev=x_list[i]   #x_i
            y_prev=y_list[i]    #y_i
            x_next=x_list[i+1]    #x_i+1
            y_next=y_list[i+1]   #y_i+1
            
            y=(y_next-y_prev)/(x_next-x_prev)*(x-x_prev)+y_prev
            return y
        



x_interpol=np.linspace(min(x_list),max(x_list),1000000)
y_interpol=[]
for i in range(0,len(x_interpol)):
    y_interpol.append(InterpoladorLineal(x_list,y_list,x_interpol[i]))
    
#print(x_interpol)
#print(InterpoladorLineal(x_list,y_list,7.5))
```
